**Exercício 6:** Considere fibras de índice-gradual com perfis de índice de $\alpha=2.0$, índice de refração da casca $n_2=1.50$ e diferenças de índice $\Delta=0.01$. Usando a Equação (7), faça o gráfico da razão $M_{eff}/M_\infty$ para raios de curvatura menores que 10 cm em $\lambda=1 \mu m$ para fibras com núcleos de raios de 4, 25 e 100 $\mu m$

```python
# Importando bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
# Definindo função para cálculo de mef
def calc_mef(alpha, delta, a, R, n2, k):
    m1 = (alpha + 2) / (2 * alpha * delta)
    m2 = (2 * a) / R
    m3 = (3 / (2 * n2 * k * R)) ** (2/3)
    mef = 1 - (m1 * (m2 + m3))   
    return mef
```

```python
# Inicializando parâmetros
alpha = 2
n2 = 1.5
delta = 0.01
lamda = 1e-6

# Curvatura de raios e parâmetro k
R = np.arange(0, 0.1, 0.01)
k = 2 * np.pi / lamda

# Raios do núcleo das fibras
a_list = np.array([4, 25, 100]) * 1e-6
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
for a in a_list:
    mef = calc_mef(alpha, delta, a, R, n2, k)
    data_dict = {}
    data_dict['mef'] = mef
    data_dict['R'] = R
    data = pd.DataFrame(data_dict)
    sns.lineplot(x=R, y=mef, data=data, label=f'a={a}')
plt.title('Curva Mef / M para variações do raio do núcleo', size=14, color='dimgrey')
plt.show()
```
